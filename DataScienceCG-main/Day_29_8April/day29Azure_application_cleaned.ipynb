```
# Scenario:
# You are a data science trainer preparing a demo for a group of first-time managers in e-commerce.
# You want to show them how to securely connect to Azure OpenAI from Google Colab, send a simple query,
#  and interpret the response. The demo emphasizes security (using secrets) and clarity (structured output).


from google.colab import userdata
from openai import OpenAI

# 🔐 Load API Key securely from Colab Secrets
api_key = userdata.get("OPENAI_API_KEY")

# ✅ Azure resource endpoint and deployment
# Correct the endpoint to be just the base URL of the Azure OpenAI resource
endpoint = "https://aryanprojectone-resource.openai.azure.com/"
deployment_name = "gpt-oss-120b"

# Initialize client with proper Azure endpoint and API version
client = OpenAI(
    api_key=api_key,
    # Construct base_url correctly: resource_base_url + /openai/deployments/ + deployment_name + /
    base_url=f"{endpoint}openai/deployments/{deployment_name}/",
    default_query={"api-version": "2024-02-01"}  # required for Azure
)

# Create a chat completion
completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "What is the capital of India?",
        }
    ],
)

# ✅ Access the message content correctly
print(completion.choices[0].message.content)
```

#Scenario
You’re training healthcare analysts.
You want them to see how to connect securely to Azure OpenAI, ask questions about AI in healthcare, and then deploy the interaction into a Gradio interface so they can try it themselves.

```
import gradio as gr

def ask_ai_healthcare(question):
    try:
        completion = client.chat.completions.create(
            model=deployment_name,
            messages=[
                {
                    "role": "user",
                    "content": "Please provide suggestions or plain text answers, avoiding tables, to the following question: " + question,
                }
            ],
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}"
```

```
import gradio as gr

with gr.Blocks(title="AI in Healthcare Assistant (powered by Azure OpenAI)") as demo:
    gr.Markdown("# AI in Healthcare Assistant (powered by Azure OpenAI)")
    gr.Markdown("Ask any question related to AI in healthcare and get a response from Azure OpenAI.")

    with gr.Column(variant="panel"):
        question_input = gr.Textbox(lines=5, placeholder="e.g., 'How can AI assist in early disease detection?'", label="Your Question")
        submit_button = gr.Button("Get AI Response")

    with gr.Column(variant="panel"):
        output_text = gr.Textbox(label="AI's Answer", interactive=False, lines=10)

    submit_button.click(
        fn=ask_ai_healthcare,
        inputs=[question_input],
        outputs=[output_text]
    )

demo.launch(debug=False, share=True)
```
