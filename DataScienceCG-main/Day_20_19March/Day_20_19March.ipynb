```python
# 🎯 Example 1: Sequential Multi-Agent System (Pipeline)
# 🧠 Scenario

# “A travel company has different employees (agents):

# Planner → decides steps

# Flight Agent → finds flights

# Weather Agent → checks weather

# Decision Agent → gives final answer”

# ================================
# AGENT 1: PLANNER
# ================================
def planner_agent(user_query):
    print("\n[Planner Agent] Creating plan...")
    return ["flight", "weather", "decision"]


# ================================
# AGENT 2: FLIGHT AGENT
# ================================
def flight_agent():
    print("\n[Flight Agent] Fetching flights...")
    return [
        {"airline": "IndiGo", "price": 4500},
        {"airline": "Air India", "price": 5200}
    ]


# ================================
# AGENT 3: WEATHER AGENT
# ================================
def weather_agent():
    print("\n[Weather Agent] Checking weather...")
    return {"condition": "Clear", "temp": 28}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(flights, weather):
    print("\n[Decision Agent] Making decision...")

    cheapest = min(flights, key=lambda x: x["price"])

    if weather["condition"] == "Rain":
        return "Avoid travel due to bad weather"

    return f"Book {cheapest['airline']} at ₹{cheapest['price']}"


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def travel_multi_agent(user_query):
    print("User Query:", user_query)

    plan = planner_agent(user_query)

    flights = None
    weather = None

    for step in plan:
        if step == "flight":
            flights = flight_agent()

        elif step == "weather":
            weather = weather_agent()

        elif step == "decision":
            result = decision_agent(flights, weather)

    return result


# RUN
response = travel_multi_agent("Plan my trip Delhi to Mumbai")
print("\nFinal Answer:", response)
```

```python
# Example 2: Manager–Worker Multi-Agent System
# 🧠 Scenario

# “Now instead of fixed flow, we introduce a Manager Agent
# that assigns tasks dynamically to worker

# ================================
# WORKER AGENTS
# ================================
def flight_agent():
    print("[Flight Agent] Working...")
    return [{"airline": "IndiGo", "price": 4500},
            {"airline": "Air India", "price": 5200}]


def weather_agent():
    print("[Weather Agent] Working...")
    return {"condition": "Clear", "temp": 28}


# ================================
# MANAGER AGENT
# ================================
def manager_agent(user_query):
    print("\n[Manager Agent] Analyzing task...")

    tasks = []

    if "trip" in user_query.lower():
        tasks = ["flight", "weather"]

    return tasks


# ================================
# EXECUTION
# ================================
def run_system(user_query):
    print("User Query:", user_query)

    tasks = manager_agent(user_query)

    results = {}

    for task in tasks:
        if task == "flight":
            results["flights"] = flight_agent()

        elif task == "weather":
            results["weather"] = weather_agent()

    # Final decision
    cheapest = min(results["flights"], key=lambda x: x["price"])

    return f"Manager Decision: Book {cheapest['airline']} at ₹{cheapest['price']}"


# RUN
response = run_system("Plan my trip")
print("\nFinal Answer:", response)
```

Scenario
“A hospital uses different employees (agents) to handle patient care in sequence.”

================================
AGENT 1: Intake Agent (Planner)
- Collects patient symptoms and history
- Decides which steps are needed (tests, consultations, etc.)

================================
AGENT 2: Diagnostic Agent
- Orders lab tests or scans
- Interprets results and identifies possible conditions

================================
AGENT 3: Treatment Agent
- Suggests treatment options (medication, therapy, surgery)
- Considers patient preferences and medical guidelines

================================
AGENT 4: Decision Agent
- Reviews all inputs (history, diagnostics, treatment options)
- Provides the final recommendation to the patient

```python
# -------------------------------
# AGENT 1: Intake Agent
# -------------------------------
class IntakeAgent:
    def process(self, patient_data):
        print("Intake Agent: Collecting patient data...")

        symptoms = patient_data.get("symptoms")
        history = patient_data.get("history")

        plan = {
            "symptoms": symptoms,
            "history": history,
            "next_step": "diagnosis"
        }

        return plan


# -------------------------------
# AGENT 2: Diagnostic Agent
# -------------------------------
class DiagnosticAgent:
    def process(self, intake_data):
        print("Diagnostic Agent: Analyzing symptoms...")

        symptoms = intake_data["symptoms"]

        # Simple rule-based diagnosis
        if "fever" in symptoms and "cough" in symptoms:
            diagnosis = "Flu"
        elif "chest pain" in symptoms:
            diagnosis = "Heart Issue"
        else:
            diagnosis = "General Checkup"

        result = {
            "diagnosis": diagnosis,
            "next_step": "treatment"
        }

        return result


# -------------------------------
# AGENT 3: Treatment Agent
# -------------------------------
class TreatmentAgent:
    def process(self, diagnosis_data):
        print("Treatment Agent: Suggesting treatment...")

        diagnosis = diagnosis_data["diagnosis"]

        if diagnosis == "Flu":
            treatment = "Rest + Paracetamol"
        elif diagnosis == "Heart Issue":
            treatment = "Consult Cardiologist"
        else:
            treatment = "Basic Medication"

        result = {
            "treatment": treatment,
            "next_step": "decision"
        }

        return result


# -------------------------------
# AGENT 4: Decision Agent
# -------------------------------
class DecisionAgent:
    def process(self, intake, diagnosis, treatment):
        print("Decision Agent: Final decision...")

        final_output = {
            "Symptoms": intake["symptoms"],
            "Diagnosis": diagnosis["diagnosis"],
            "Treatment": treatment["treatment"]
        }

        return final_output


# -------------------------------
# PIPELINE EXECUTION
# -------------------------------
def run_pipeline(patient_data):
    intake_agent = IntakeAgent()
    diagnostic_agent = DiagnosticAgent()
    treatment_agent = TreatmentAgent()
    decision_agent = DecisionAgent()

    intake_output = intake_agent.process(patient_data)
    diagnosis_output = diagnostic_agent.process(intake_output)
    treatment_output = treatment_agent.process(diagnosis_output)

    final_result = decision_agent.process(
        intake_output,
        diagnosis_output,
        treatment_output
    )

    return final_result


# -------------------------------
# TEST
# -------------------------------
patient = {
    "symptoms": ["fever", "cough"],
    "history": "No major illness"
}

result = run_pipeline(patient)

print("\nFinal Recommendation:")
print(result)
```

Scenario: Corporate Market Research & Strategy
A company wants to explore launching a new product in a competitive market. The Manager Agent oversees the process and delegates tasks to specialized workers.

👩‍💼 Manager Agent
- Receives the overall goal: “Evaluate feasibility of launching Product Y in Asia.”
- Dynamically assigns tasks to worker agents depending on what’s needed.
- Example: If budget is unclear → send to Finance Worker. If regulations are complex → send to Legal Worker.

================================
📊 Worker Agents
================================
- Market Research Worker
- Collects competitor data, customer preferences, and demand forecasts.
- Reports: “High demand in Tier‑1 cities, moderate competition.”
- Finance Worker
- Analyzes budget, ROI, and pricing strategy.
- Reports: “Estimated $10M investment, ROI in 2 years.”
- Operations Worker
- Evaluates supply chain, production capacity, and logistics.
- Reports: “Factories can scale up, but shipping costs are high.”
- Legal Worker
- Reviews compliance, intellectual property, and regional regulations.
- Reports: “Trademark available, but import laws require certification.”
- HR Worker
- Assesses staffing needs and training requirements.
- Reports: “Need 50 new hires for customer support and sales.”

```python
# ================================
# Corporate Market Research System
# ================================

# -------- Worker Agents --------

# Without API

class MarketResearchWorker:
    def run(self):
        return {
            "agent": "Market Research",
            "report": "High demand in Tier-1 cities, moderate competition."
        }


class FinanceWorker:
    def run(self):
        return {
            "agent": "Finance",
            "report": "Estimated $10M investment, ROI expected in 2 years."
        }


class OperationsWorker:
    def run(self):
        return {
            "agent": "Operations",
            "report": "Factories can scale up, but shipping costs are high."
        }


class LegalWorker:
    def run(self):
        return {
            "agent": "Legal",
            "report": "Trademark available, but import laws require certification."
        }


class HRWorker:
    def run(self):
        return {
            "agent": "HR",
            "report": "Need 50 new hires for customer support and sales."
        }


# -------- Manager Agent --------

class ManagerAgent:
    def __init__(self, goal):
        self.goal = goal
        self.reports = []

        # Initialize workers
        self.market = MarketResearchWorker()
        self.finance = FinanceWorker()
        self.operations = OperationsWorker()
        self.legal = LegalWorker()
        self.hr = HRWorker()

    def decide_tasks(self):
        print(f"\n🎯 Goal: {self.goal}")

        # Dynamic decision logic
        tasks = []

        if "market" in self.goal.lower() or "product" in self.goal.lower():
            tasks.append(self.market)

        if "budget" in self.goal.lower() or "feasibility" in self.goal.lower():
            tasks.append(self.finance)

        if "launch" in self.goal.lower():
            tasks.append(self.operations)

        if "asia" in self.goal.lower():
            tasks.append(self.legal)

        # HR is always needed for new product
        tasks.append(self.hr)

        return tasks

    def execute(self):
        workers = self.decide_tasks()

        print("\n👩‍💼 Manager assigning tasks...\n")

        for worker in workers:
            result = worker.run()
            self.reports.append(result)
            print(f"✅ {result['agent']} Worker Report:")
            print(f"   ➤ {result['report']}\n")

        return self.final_decision()

    def final_decision(self):
        print("\n📌 Final Decision by Manager:\n")

        # Simple decision logic
        decision = "Product launch is feasible with moderate risk.\n"

        # Add insights
        for report in self.reports:
            decision += f"- {report['agent']}: {report['report']}\n"

        decision += "\n👉 Recommendation: Proceed with launch in Tier-1 cities with cost optimization."

        return decision


# -------- MAIN --------

if __name__ == "__main__":
    goal = "Evaluate feasibility of launching Product Y in Asia"

    manager = ManagerAgent(goal)
    final_output = manager.execute()

    print(final_output)
```

Scenario: Corporate Product Launch Broadcast
Imagine a company preparing to launch Product X in Q3. The Coordinator Agent (like a corporate program manager) sends out a broadcast announcement to all departments at once:
“Product X launch in Q3, target market North America, budget $5M.”


📢 Coordinator Agent (Broadcaster)

- This is implemented in the code by the broadcast() function, which uses asyncio.gather() to run all agents in parallel.

📈 Marketing Agent
- Receives the broadcast.
- Responds with a marketing strategy: campaigns, channels, and positioning.
💰 Finance Agent
- Receives the broadcast.
- Responds with budget allocation and ROI forecasts.

- Responds with staffing and training plans.

- Collects all responses.

- In the code, this is the decision_agent() function that merges all outputs into one consolidated plan.

```python
import requests
from google.colab import userdata

# ---------------- API KEY ---------------- #
# Store your Groq API key in Colab Secrets:
# Click 🔑 (key icon) in the left sidebar → Add secret named: groq_api_key
API_KEY = userdata.get('API_KEY')

if not API_KEY:
    raise ValueError("❌ Groq API key not found. Add 'groq_api_key' in Colab Secrets (🔑 icon in sidebar).")

print("✅ Groq API key loaded.")

# ---------------- LLM CALL ---------------- #
def call_llm(prompt):
    url = "https://api.groq.com/openai/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    data = {
        "model": "llama-3.1-8b-instant",
        "messages": [
            {"role": "system", "content": "You are a precise business analyst."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.3,
        "max_tokens": 800
    }
    try:
        res = requests.post(url, headers=headers, json=data, timeout=30)
        if res.status_code != 200:
            print("Status:", res.status_code)
            print("Response:", res.text)
        res.raise_for_status()
        return res.json()["choices"][0]["message"]["content"]
    except Exception as e:
        return f"Error: {e}"


# ---------------- AGENTS ---------------- #
class Agent:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def respond(self, message):
        print(f"\n[{self.name}] thinking...")
        prompt = f"""You are {self.name}.
Role: {self.role}

Message: {message}

Give 2-3 short actionable steps."""
        result = call_llm(prompt)
        print(f"[{self.name}] {result}")
        return result


class DecisionAgent:
    def decide(self, responses):
        print("\n[Decision Agent] combining responses...")
        combined = "\n".join(responses)
        prompt = f"Combine these expert inputs into one concise final plan:\n\n{combined}"
        result = call_llm(prompt)
        print("\n===== FINAL PLAN =====")
        print(result)


class Coordinator:
    def __init__(self, agents):
        self.agents = agents

    def broadcast(self, message):
        print(f"\n[Coordinator] Broadcasting: {message}")
        return [agent.respond(message) for agent in self.agents]


# ---------------- RUN ---------------- #
if __name__ == "__main__":
    message = "Product X launch in Q3, North America, budget $5M"

    agents = [
        Agent("Marketing Agent", "marketing strategy"),
        Agent("Finance Agent", "budget planning"),
        Agent("Operations Agent", "supply chain"),
        Agent("Legal Agent", "compliance"),
        Agent("HR Agent", "staffing")
    ]

    coordinator = Coordinator(agents)
    decision = DecisionAgent()

    responses = coordinator.broadcast(message)
    decision.decide(responses)
```

```python
# ==========================================
# AI RESEARCH TEAM PIPELINE (GROQ API)
# ==========================================

import requests
from google.colab import userdata

# 🔑 Load API Key from Colab Secrets
API_KEY = userdata.get('groq_api_key')

if not API_KEY:
    raise ValueError("❌ API key not found in Colab secrets")

print("✅ API key loaded:", API_KEY[:10], "...")


# ---------------- LLM CALL ---------------- #
def call_llm(prompt):
    url = "https://api.groq.com/openai/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "llama-3.1-8b-instant",  # fast + free model
        "messages": [
            {"role": "system", "content": "You are a precise business analyst."},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.3,
        "max_tokens": 800
    }

    try:
        res = requests.post(url, headers=headers, json=data, timeout=30)

        if res.status_code != 200:
            print("❌ Status:", res.status_code)
            print("Response:", res.text)

        res.raise_for_status()
        return res.json()["choices"][0]["message"]["content"]

    except Exception as e:
        return f"Error: {e}"


# ---------------- AGENTS ---------------- #

class SearchAgent:
    def run(self, goal):
        print("\n🔍 [Search Agent] collecting data...")

        prompt = f"""
Goal: {goal}

Find key facts, statistics, and recent trends about EV industry in India (focus 2025).
Keep it concise.
"""
        return call_llm(prompt)


class AnalystAgent:
    def run(self, data):
        print("\n📊 [Analyst Agent] analyzing...")

        prompt = f"""
Analyze the following EV market data:

{data}

Extract:
- Trends
- Growth drivers
- Challenges
- Key insights
"""
        return call_llm(prompt)


class WriterAgent:
    def run(self, analysis):
        print("\n✍️ [Writer Agent] writing report...")

        prompt = f"""
Create a professional market report based on:

{analysis}

Structure:
- Introduction
- Market Trends
- Challenges
- Opportunities
- 2025 Outlook
"""
        return call_llm(prompt)


class QAAgent:
    def run(self, report):
        print("\n✅ [QA Agent] reviewing report...")

        prompt = f"""
Review and improve this report:

{report}

Ensure:
- Clarity
- Accuracy
- Professional tone
"""
        return call_llm(prompt)


# ---------------- ORCHESTRATOR ---------------- #

class Orchestrator:
    def run(self, goal):
        print(f"\n🎯 [Orchestrator] Goal: {goal}")

        search = SearchAgent()
        analyst = AnalystAgent()
        writer = WriterAgent()
        qa = QAAgent()

        data = search.run(goal)
        analysis = analyst.run(data)
        draft = writer.run(analysis)
        final = qa.run(draft)

        return final


# ---------------- RUN ---------------- #

if __name__ == "__main__":
    goal = "Write a comprehensive market report on EV industry trends in India for 2025"

    orchestrator = Orchestrator()
    final_report = orchestrator.run(goal)

    print("\n\n================ FINAL REPORT ================\n")
    print(final_report)
```
