```python
# Scenario: AI Content Automation for a Startup
# 🏢 Problem Statement

# A startup wants to:

# Stay updated with latest AI trends

# Automatically create LinkedIn posts

# Save time on manual research & writing

# 👉 So we build a multi-agent AI system to automate this using CrewAI.

!pip install crewai litellm langchain-community langchain-litellm

import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM # Updated import

# -----------------------------
# 1. Set Groq API Key
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("groq_api_key")

# ❌ REMOVE any OPENAI key completely
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. Define LLM (using ChatLiteLLM for Groq)
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------
researcher = Agent(
    role="Researcher",
    goal="Find latest AI news",
    backstory="Expert in AI trends",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Write LinkedIn post",
    backstory="Expert content creator",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------
research_task = Task(
    description="Find 5 latest AI news trends",
    expected_output="List with summaries",
    agent=researcher
)

write_task = Task(
    description="Write LinkedIn post",
    expected_output="Final post",
    agent=writer,
    context=[research_task]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
```

Scenario: Automated Learning Hub for an EdTech Company
🎓 Problem Statement
An EdTech company wants to:
- Keep track of the latest educational technology innovations
- Automatically generate weekly blog posts for their website
- Reduce time spent on manual research and content drafting
👉 So we build a multi-agent AI system using CrewAI to automate this workflow:
- Trend Monitoring Agent → Continuously scans for new EdTech tools, teaching strategies, and AI-in-education updates
- Content Summarization Agent → Condenses research into digestible insights tailored for educators and learners
- Blog Writing Agent → Crafts engaging, SEO-friendly blog posts automatically
- Publishing Agent → Schedules and posts content directly to the company’s website

```python
import os
from crewai import Agent, Task, Crew
from langchain_litellm import ChatLiteLLM

# Load environment variables
load_dotenv()
os.environ["GROQ_API_KEY"] = userdata.get("groq_api_key")

# LLM configuration
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")


# -------------------------
# Agents
# -------------------------

trend_monitor = Agent(
    role="EdTech Trend Analyst",
    goal="Identify latest innovations in education technology",
    backstory="Expert in EdTech tools, AI learning platforms, and modern teaching methods",
    llm=llm,
    verbose=True
)

summarizer = Agent(
    role="Education Content Analyst",
    goal="Summarize trends into insights useful for teachers and learners",
    backstory="Expert in simplifying complex EdTech developments",
    llm=llm,
    verbose=True
)

writer = Agent(
    role="SEO Blog Writer",
    goal="Write engaging educational blog posts",
    backstory="Expert in writing blogs optimized for search engines",
    llm=llm,
    verbose=True
)

publisher = Agent(
    role="Content Publisher",
    goal="Prepare blog for publishing",
    backstory="Ensures formatting, readability, and publishing schedule",
    llm=llm,
    verbose=True
)

# -------------------------
# Tasks
# -------------------------

trend_task = Task(
    description="""
    Find latest trends in EdTech:
    - AI tutors
    - adaptive learning
    - gamification tools
    - virtual classrooms
    - skill-based learning platforms
    """,
    expected_output="List of 5 trends with short explanation",
    agent=trend_monitor
)

summary_task = Task(
    description="""
    Summarize trends into simple insights.
    Focus on:
    - how it helps students
    - how teachers can use it
    """,
    expected_output="Concise structured summary",
    agent=summarizer,
    context=[trend_task]
)

blog_task = Task(
    description="""
    Write SEO-friendly blog post:
    Title
    Introduction
    5 sections explaining trends
    conclusion
    keywords
    """,
    expected_output="Complete blog article",
    agent=writer,
    context=[summary_task]
)

publish_task = Task(
    description="""
    Format blog professionally:
    - headings
    - bullet points
    - add tags
    - suggest publishing schedule
    """,
    expected_output="Final ready-to-publish blog",
    agent=publisher,
    context=[blog_task]
)

# -------------------------
# Crew
# -------------------------

crew = Crew(
    agents=[trend_monitor, summarizer, writer, publisher],
    tasks=[trend_task, summary_task, blog_task, publish_task],
    verbose=True
)

# Run system
result = crew.kickoff()

print("\n FINAL BLOG:\n")
print(result)
```

```python
# Scenario: Smart Job Application Assistant
# 🏢 Problem Statement

# Students struggle to:

# Analyze job descriptions

# Match their skills

# Write tailored resumes & cover letters

# 👉 So we build a multi-agent AI system that automates job application preparation.

# 🤖 System Overview
# 🔹 Agent 1: Job Analyzer

# Reads job description

# Extracts required skills

# 🔹 Agent 2: Resume Optimizer

# Matches candidate skills with job

# Suggests improvements

# 🔹 Agent 3: Cover Letter Writer

# Generates a personalized cover letter

# 🔄 Workflow
# Job Description → Skill Analysis → Resume Optimization → Cover Letter Generation


!pip install crewai litellm

import os
from crewai import Agent, Task, Crew
from google.colab import userdata

# -----------------------------
# 1. API Key Setup
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("groq_api_key")

# -----------------------------
# 2. Define Agents
# -----------------------------
job_analyzer = Agent(
    role="Job Analyzer",
    goal="Extract key skills and requirements from job descriptions",
    backstory=(
        "You are an HR expert who understands job postings and identifies "
        "important skills, qualifications, and expectations."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

resume_optimizer = Agent(
    role="Resume Optimizer",
    goal="Align candidate resume with job requirements",
    backstory=(
        "You are a career coach who improves resumes by matching them "
        "with job descriptions and highlighting relevant skills."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

cover_letter_writer = Agent(
    role="Cover Letter Writer",
    goal="Write a personalized cover letter",
    backstory=(
        "You are a professional writer who crafts compelling cover letters "
        "tailored to specific job roles."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Input Data (Scenario)
# -----------------------------
job_description = """
We are looking for a Data Analyst with experience in Python, SQL, Power BI,
and data visualization. The candidate should be able to analyze datasets,
create dashboards, and communicate insights effectively.
"""

candidate_resume = """
I am a BTech student skilled in Python and basic SQL.
I have created Power BI dashboards and completed projects on data analysis.
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following job description and extract:
    - Required skills
    - Key responsibilities
    - Tools/technologies

    Job Description:
    {job_description}
    """,
    expected_output="Structured list of skills, responsibilities, and tools",
    agent=job_analyzer
)

resume_task = Task(
    description=f"""
    Compare the candidate resume with the job requirements and:
    - Identify skill gaps
    - Suggest improvements
    - Optimize the resume

    Resume:
    {candidate_resume}
    """,
    expected_output="Improved resume with suggestions",
    agent=resume_optimizer,
    context=[analysis_task]
)

cover_letter_task = Task(
    description="""
    Create a personalized cover letter using:
    - Job requirements
    - Candidate resume
    - Improvements suggested

    Make it professional and engaging.
    """,
    expected_output="Final cover letter",
    agent=cover_letter_writer,
    context=[analysis_task, resume_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[job_analyzer, resume_optimizer, cover_letter_writer],
    tasks=[analysis_task, resume_task, cover_letter_task],
    verbose=True,
    tracing=True   # good for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n🚀 FINAL OUTPUT:\n")
print(result)
```

```python
# Scenario: Automated Business Intelligence (BI) Report Generator
# 🎯 Problem Statement

# In companies:

# Managers struggle to analyze large sales data

# Reports take hours to prepare

# Insights are often delayed

# 👉 Solution: Build a multi-agent AI system that:

# Analyzes business data

# Extracts insights

# Generates executive reports

# 🤖 System Overview
# 🔹 Agent 1: Data Analyst

# Analyzes sales/business data

# Finds trends, patterns, KPIs

# 🔹 Agent 2: Business Strategist

# Interprets insights

# Suggests business decisions

# 🔹 Agent 3: Report Generator

# Creates a professional executive report


!pip install crewai litellm

import os
from crewai import Agent, Task, Crew
from google.colab import userdata

# -----------------------------
# 1. API Key Setup
# -----------------------------

os.environ["GROQ_API_KEY"] = userdata.get("groq_api_key")

# -----------------------------
# 2. Define Agents
# -----------------------------
data_analyst = Agent(
    role="Data Analyst",
    goal="Analyze business data and extract meaningful insights",
    backstory=(
        "You are a senior data analyst working in a corporate environment. "
        "You specialize in identifying trends, KPIs, and patterns in sales data."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

business_strategist = Agent(
    role="Business Strategist",
    goal="Provide strategic recommendations based on data insights",
    backstory=(
        "You are a business consultant who translates data insights into "
        "actionable strategies for company growth."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

report_generator = Agent(
    role="Executive Report Writer",
    goal="Generate a professional business report",
    backstory=(
        "You are an expert in corporate communication who prepares executive-level "
        "reports with clarity and professionalism."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Sample Business Data
# -----------------------------
sales_data = """
Region: North | Revenue: 12000 | Cost: 8000
Region: South | Revenue: 15000 | Cost: 9000
Region: East  | Revenue: 10000 | Cost: 7000
Region: West  | Revenue: 18000 | Cost: 11000
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following sales data and provide:
    - Total revenue and profit
    - Best performing region
    - Key trends and observations

    Data:
    {sales_data}
    """,
    expected_output="Detailed analysis with KPIs and trends",
    agent=data_analyst
)

strategy_task = Task(
    description="""
    Based on the analysis:
    - Suggest 3 business strategies
    - Identify growth opportunities
    - Highlight risks
    """,
    expected_output="Strategic recommendations",
    agent=business_strategist,
    context=[analysis_task]
)

report_task = Task(
    description="""
    Create a professional executive report including:
    - Summary
    - Key insights
    - Recommendations
    - Conclusion
    """,
    expected_output="Well-structured business report",
    agent=report_generator,
    context=[analysis_task, strategy_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[data_analyst, business_strategist, report_generator],
    tasks=[analysis_task, strategy_task, report_task],
    verbose=True,
    tracing=True   # great for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n📊 FINAL BUSINESS REPORT:\n")
print(result)
```

 Scenario: Smart Sales Proposal Assistant
📌 Problem Statement
Enterprise sales teams often struggle to:
- Analyze client requirements from RFPs (Requests for Proposals)
- Match company offerings to client needs
- Draft tailored proposals that stand out
- Ensure consistency across documents
👉 So we build a multi-agent AI system that automates sales proposal preparation.

🤖 System Overview
- 🔹 Agent 1: RFP Analyzer
- Reads client’s request for proposal
- Extracts key requirements, budget constraints, and evaluation criteria
- 🔹 Agent 2: Solution Mapper
- Matches company products/services to client needs
- Suggests tailored value propositions and differentiators
- 🔹 Agent 3: Proposal Writer
- Generates a polished, client-specific proposal
- Ensures alignment with brand tone and compliance standards

🔄 Workflow
RFP Document → Requirement Analysis → Solution Mapping → Proposal Generation

```python
!pip install crewai litellm langchain-litellm
```

```python
import os
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM

# -----------------------------
# 1. API Key Setup
# -----------------------------
os.environ["GROQ_API_KEY"] = userdata.get("groq_api_key")

# -----------------------------
# 2. Define LLM (using ChatLiteLLM for Groq)
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------
rfp_analyzer = Agent(
    role="RFP Analyzer",
    goal="Extract key requirements, budget constraints, and evaluation criteria from client RFPs",
    backstory=(
        "You are an expert business analyst specializing in dissecting Request for Proposals (RFPs). "
        "Your goal is to meticulously identify core client needs, budget limitations, and critical "
        "evaluation metrics to guide proposal development."
    ),
    llm=llm,
    verbose=True
)

solution_mapper = Agent(
    role="Solution Mapper",
    goal="Match company offerings to client needs and suggest tailored value propositions",
    backstory=(
        "You are a strategic solutions architect with deep knowledge of company products/services. "
        "Your expertise lies in aligning our capabilities with specific client requirements, "
        "crafting compelling value propositions, and identifying key differentiators."
    ),
    llm=llm,
    verbose=True
)

proposal_writer = Agent(
    role="Proposal Writer",
    goal="Generate a polished, client-specific sales proposal, ensuring brand tone and compliance",
    backstory=(
        "You are a professional proposal writer known for crafting persuasive and coherent sales documents. "
        "You ensure proposals are tailored, resonate with the client's needs, and adhere to all "
        "branding and compliance standards."
    ),
    llm=llm,
    verbose=True
)

# -----------------------------
# 4. Input Data (Scenario)
# -----------------------------
rfp_document = """
Client: Tech Solutions Inc.
Project: Cloud Migration Services

Overview:
Tech Solutions Inc. is seeking a partner to assist with a comprehensive cloud migration from on-premise infrastructure to a hybrid cloud environment. We require expertise in AWS and Azure, with a focus on cost optimization, security, and minimal downtime during migration.

Key Requirements:
- Experience with large-scale enterprise cloud migrations (over 500 VMs).
- Proficiency in both AWS and Azure cloud platforms.
- Strong focus on data security and compliance (GDPR, ISO 27001).
- Proven strategies for cost optimization in cloud environments.
- Disaster recovery planning and implementation.
- Post-migration support and optimization.

Budget: Up to $500,000
Timeline: 6 months
Evaluation Criteria:
- Technical expertise (35%)
- Proposed methodology (30%)
- Cost-effectiveness (20%)
- Previous experience/case studies (15%)
"""

company_offerings = """
Our Company: Global Cloud Solutions

Services:
- Full-spectrum cloud migration (AWS, Azure, GCP).
- Cloud architecture design and optimization.
- Managed security services (ISO 27001 certified, GDPR compliant).
- Cost management and FinOps consulting.
- Business continuity and disaster recovery solutions.
- 24/7 post-migration support.

Key Differentiators:
- 10+ years of experience in enterprise cloud transformations.
- Proprietary AI-driven cost optimization tools.
- Dedicated security compliance team.
- Successful migrations for clients like "Enterprise Bank" and "Healthcare Systems Ltd."
"""

# -----------------------------
# 5. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following RFP document to extract:
    - Client Name
    - Project Scope
    - Key Technical Requirements (e.g., cloud platforms, security needs)
    - Non-Technical Requirements (e.g., timeline, budget)
    - Evaluation Criteria and their weightings

    RFP Document:
    {rfp_document}
    """,
    expected_output="A structured list of all extracted RFP details, categorized for clarity.",
    agent=rfp_analyzer
)

mapping_task = Task(
    description=f"""
    Based on the RFP analysis and our company's offerings, do the following:
    - Identify which of 'Global Cloud Solutions' services directly address each 'Tech Solutions Inc.' requirement.
    - Highlight our key differentiators that would be most impactful for this client.
    - Suggest specific value propositions tailored to their needs (e.g., how our cost optimization tools save them money).

    Company Offerings:
    {company_offerings}
    """,
    expected_output="A detailed mapping of company strengths to client needs, including tailored value propositions and differentiators.",
    agent=solution_mapper,
    context=[analysis_task]
)

proposal_task = Task(
    description=f"""
    Generate a professional sales proposal for 'Tech Solutions Inc.' based on the RFP analysis and solution mapping.
    The proposal should include:
    - An executive summary.
    - A detailed section addressing each of their key requirements with our corresponding solution.
    - A section on our unique value propositions and differentiators.
    - A brief overview of our experience (mentioning case studies if relevant).
    - A conclusion with a call to action.
    - Ensure the tone is professional, confident and persuasive.
    - Adhere to the evaluation criteria by emphasizing relevant points.

    RFP Document for context:
    {rfp_document}

    Company Offerings for context:
    {company_offerings}
    """,
    expected_output="A complete, well-structured, and persuasive sales proposal ready for submission.",
    agent=proposal_writer,
    context=[analysis_task, mapping_task]
)

# -----------------------------
# 6. Crew Setup
# -----------------------------
crew = Crew(
    agents=[rfp_analyzer, solution_mapper, proposal_writer],
    tasks=[analysis_task, mapping_task, proposal_task],
    verbose=True,
    tracing=True
)

# -----------------------------
# 7. Run
# -----------------------------
result = crew.kickoff()

print("\n📝 FINAL SALES PROPOSAL:\n")
print(result)
```
